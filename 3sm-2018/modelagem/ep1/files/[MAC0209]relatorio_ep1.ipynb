### 1. Introdução
   <p> O movimento retilíneo é caracterizado pelo deslocamento em trajetórias retas. Dentro desse assunto, duas classificações serão abordadas por esse experimento: o Movimento Retilíneo Uniforme (MRU) e o Movimento Retilíneo Uniformemene Variado (MRUV). 
    </p> 
    <p>No caso do primeiro, trata-se de uma movimentação sem a aceleração, com velocidade constante. O segundo, por sua vez, contém aceleração constante, porém, velocidade que muda.
    </p>
    <p>Apesar de buscar-se reproduzir fielmente esses movimentos, no mundo real, existem outras variáveis que são mais difíceis de controlar, como por exemplo, atrito com o chão ou até trajetórias que não são exatamente retas. Essas variáveis extrínsecas acabam gerando algumas distorções nos resultados, como será possível verificar mais adiante. 
    </p>
    <p>Estudado desde pelo menos a Revolução Científica, por volta do século XVI, os movimentos retilíneos estão presentes em muitos aspectos do nosso dia-a-dia. A sua reprodução com um tratamento computacional é extremamente importante, pois possibilita análises mais apuradas ou sobre maior quantidade de dados através das ferramentas de <i>e-science</i> disponíveis. E é exatamente sobre essa abordagem que esse relatório do experimento se voltará.
    </p> <br>

**2.2 Método**
       <p> Para realização da reprodução do MRU, foi feita a medida do corredor com a trena e marcadas a giz as posições 0m, 5m, 10m, 15m, 20m, 25m e 30m. Primeiramente, cogitou-se usar um barbante amarrado entre as pernas. No entanto, percebeu-se que isso demandaria um certo tempo para se repetir nas demais pessoas que fariam o experimento. Portanto, optou-se pelo uso do metrônomo, ajustado para 65 BPM, aliado com as passadas uniformes, marcadas pela distância das pedras das quais o corredor era feito. Pareceu bastante adequado ao grupo que as medidas fossem tomadas por uma única dupla usando o cronômetro do celular, tanto nas distâncias de 10 em 10 metros quanto nas de 5 em 5. 
       </p>
       <p> Já no MRUV, os passos de preparação foram análogos aos descritos anteriormente. Foram visitadas algumas rampas no câmpus Cidade Universitária e estudadas suas características de comprimento e inclinação. Para a simulação, foi escolhido o declive da avenida Prof. Luciano Gualberto, por ser uma área aberta, não demandar uma autorização expressa de algum órgão, ter baixa inclinação e ter mais de 30m. 
       </p>
       <p> Primeiramente, testou-se a descida do skate sem pessoa alguma montada. Contudo, verificou-se que a velocidade adquirida era muito baixa, chegando a zero em alguns momentos. Em seguida, foi experimentada a descida com pessoas, primeiramente sentadas e, depois, deitadas. Decidiu-se que cada um desceria da forma que se sentisse mais seguro. Novamente, aqui, apesar da velocidade desenvolvida ser um pouco maior, pareceu adequado que as medidas de tempo fossem tomadas por uma única dupla. </p>
       <p> Em todos os movimentos, o transeunte estava portando um celular com o aplicativo Physics Toolbox registrando as medições do acelerômetro. Além disso, no caso do MRUV, foram usados aplicativos distintos em celular diferentes para se obter uma aproximação da inclinação do declive. </p> <br>

### 5. Análise

Duas possíveis análises se sobressaem. A primeira delas diz respeito a coerência na coleta dos dados com relação ao alinhamento entre os senhores. A segunda, sobre como esses dados representam a perspectiva teórica dos movimentos.

Em termos mensuráveis, a primeira análise é marcada pela constação de que os valores coletados nas movimentações em velocidade constante se respaldam consideravelmente melhor que no caso das movimentações aceleradas. Peguemos, por exemplo, o movimento do Artur em ambos os tipos de movimento: No que diz respeito ao MRU, o desvio padrão médio entre os sensores é bastante baixo, da ordem de $8x10^{-2}$. No MRUV, entretando, tal desvio é da ordem de $1.5x10^{-1}$, ou seja, maior que o caso MRU em mais de uma ordem de grandeza.

A segunda análise, por sua vez, se pauta sobretudo na inclinação das curvas dos movimentos comparativamente entre a que é obtida a partir dos nossos dados e a que representa suas respectivas equações horárias. Nesse caso, como no MRU a curva de velocidade é tão somente uma reta horizontal e no MRUV uma reta linear inclinada, observar a curva da variação da posição no espaço pelo tempo (S x t) - linear no MRU mas parabólica no MRUV - nos oferece mais insumos. 

Posto isso, se sobressai nessa análise o fato de que a maior discrepância nos dados obtidos pelos sensores no MRUV em comparação ao MRU, como mencionado no parágrafo anterior, reverbera nas curvas de deslocamento, que encaixam mais com a previsão teórica no movimento uniforme do que nos com aceleração. Isto é, no que diz respeito ao MRU, as retas experimentais e teórica estão sempre sobrepostas, enquanto no MRUV não só não há uma sobreposição completa como também é possível perceber que a curva experimental não segue uma tendência parabólica durante todo o intervalo, adotando um compartamento quase linear em seus trechos finais em alguns casos.

Em ambos os casos, tais relações ficam mais evidentes quando se compara os movimento de 10m em 10m com os de 5m em 5m, onde esse último, pela maior quantidade de informação sobre a qual trabalhar, tende a disfarçar melhor as discrepâncias citadas, se aproximando do modelo teórico. 

### 6. Interpretação

Como dito, no MRU as medições dos dois sensores são quase coincidentes. Ademais, tirando-se a média dos dois sensores, nas três passadas, a linha da relação posição x tempo é praticamente a mesma que a da tendência linear. Tal convergência é consequência da simplicidade do descolamento, que com os devidos recursos por nós utilizado para garantir a acurácia (marcações dos espaço no chão e metrônomo), garantiram uma boa reprodução das condições ideais. É possível notar que quando são tomadas as medidas de 5 em 5 metrôs, percebemos mais variações.

Já no MRUV, há curvas fora da tendência em todas as passadas. Quanto às medições dos sensores, elas apresentam uma divergência um pouco maior. Isso pode ser explicado pelo fato de que, devido ao skate estar acelerado, os "sensores" deveriam se locomover mais rapidamente para a próxima posição para marcar o tempo, gerando essas variações. Ademais, percebe-se que as curvas que mais fogem da tendência são a do Eduardo e a do Pedro. Esse fenômeno pode estar relacionado ao fato de o Artur ser o único que possuía um maior traquejo na descida de skate. Eduardo e Pedro, às vezes tremiam o corpo, o que pode ter causado interferência na aceleração. Outro fato que pode justificar essas anomalias é a descida escolhida, que não tinha uma inclinação tão uniforme, ocasionando mais aceleração. Como dito na análise da seção anterior, em alguns casos é possível, ainda, oberservar que a curva de descolamento tem sua inclinação alterada na reta final, ficando menos parabólica e mais linear, o que corresponde a momentos onde os membros em descolamento, quando já com uma velocidade relativamente alta, acabaram por, numa reação natural involuntaria, frear um pouco o skate a fim de evitar quedas e acidentes.



### 7. Crítica
                  
Nesse experimento, o grupo pôde verificar o quão difícil é manter alguns aspectos uniformes ou constantes num ambiente não controlado, com variáveis externas que fogem do nosso alcance, por exemplo, atrito do ar e do chão. Dessa forma, entende-se que se poderia minimizar o valor do erro, mas, nas condições dessa experiência, ele sempre estaria presente. Mesmo no caso do movimento retilíneo uniformemente variado, há distorções que podem ter sido causadas pela mudança de trajetória do skate, falta de constância na inclinação, entre outros aspectos.

A importância do planejamento. Realizamos o experimento do MRU antes do MRUV apesar de sabermos que havia previsão de chuva para o horário do almoço, e que o MRUV era o único experimento que necessariamente precisava ser realizado em uma área aberta, devido a forma que resolvemos realizá-lo. Então, em certo momento, a chuva interrompeu a relização do MRUV, mas, por sorte, a tempestade deu uma trégua e conseguimos finalizar a coleta dos dados. 

Para além da questão do planejamento, outra habilidade que adquiriu-se foi quanto ao tratamento dos dados. É de suma importância definir um padrão para anotação das informações, dado que diferentes pessoas realizaram a tarefa e poderia haver divergência nos registros, por exemplo, ter anotado tempo acumulado ao invés de tempo entre cada "sensor". Ainda nesse sentido, pode-se mencionar o fato de que o grupo sabia pouco ou muito pouco sobre Python. Portanto, utilizar o Jupyter para relatar a experiência e ter incluído alguns códigos dessa linguagem no _notebook_, representou um outro ganho muito grande de conhecimento. A interfacce gráfica e facilidade que o Python e o Anaconda favorecem o manuseio dos dados, além de serem ferramentas amplamente utilizadas.

Quanto a proposição do trabalho, a única crítica negativa é quanto ao número de passadas sugeridas. Considera-se que 9 passadas (3 por pessoa), para uma única distância de 5 em 5 metros, já seria suficiente para fazer-se as devidas constatações. Apesar disso, no geral, considera-se que esse exercício represente um importante e válido treinamento para o tratamento de dados do mundo real, para que se possa extrair informações, além de uma experimentação do rigor científico para documentação da experiência.
