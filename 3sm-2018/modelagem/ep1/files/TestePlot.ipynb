```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# configuring display
pd.set_option('display.line_width', 5000) 
pd.set_option('display.max_columns', 60)    

#Objeto DataFrame recebe os dados do arquivo.
#Plotting from the collected data.
#Check if PATH is set properly.
df = pd.read_csv('../data/ep1/datasetfinal.csv', sep=',')
dfspec_1A = df['time1ArturMedia']
dfspec_2A = df['time2ArturMedia']
dfspec_3A = df['time3ArturMedia']

dfspec_1P = df['time1PedroMedia']
dfspec_2P = df['time2PedroMedia']
dfspec_3P = df['time3PedroMedia']

dfspec_1An = df['time1AndrewMedia']
dfspec_2An = df['time2AndrewMedia']
dfspec_3An = df['time3AndrewMedia']

# # Velocidade média deltas/deltat
vma = [0.7664,0.8205,0.8103]
vmp = [0.82,0.773,0.794]
vman = [0.784,0.762,0.758]
dist = [0.0, 10.0, 20.0, 30.0]
smed_1A = []
smed_2A = []
smed_3A = []

smed_1P = []
smed_2P = []
smed_3P = []

smed_1An = []
smed_2An = []
smed_3An = []

# simulating analytic formula
i = 0
while (i < 4):
    sf = dist[i] / vma[0] 
    sf_2 = dist[i] / vma[1]
    sf_3 = dist[i] / vma[2] 
    smed_1A.append(sf)
    smed_2A.append(sf_2)
    smed_3A.append(sf_3)    
    i += 1

i = 0
while (i < 4):
    sf = dist[i] / vmp[0]
    sf_2 = dist[i] / vmp[1]
    sf_3 = dist[i] / vmp[2]
    smed_1P.append(sf)
    smed_2P.append(sf_2)
    smed_3P.append(sf_3)    
    i += 1

i = 0
while (i < 4):
    sf = dist[i] / vman[0]
    sf_2 = dist[i] / vman[1]
    sf_3 = dist[i] / vman[2]
    smed_1An.append(sf)
    smed_2An.append(sf_2)
    smed_3An.append(sf_3)    
    i += 1
    
totalA = [0.0]
totalp = [0.0]
totalAn = [0.0]
for j in range(0,4):
    n = smed_1A[j]
    n_2 = smed_2A[j]
    n_3 = smed_3A[j]
    if n != 0:
        totalA.append((n + n_2 + n_3)/3.0)
    np = smed_1P[j]
    np_2 = smed_2P[j]
    np_3 = smed_3P[j]
    if np != 0:
        totalp.append((np + np_2 + np_3)/3.0)
    na = smed_1An[j]
    na_2 = smed_2An[j]
    na_3 = smed_3An[j]
    if na != 0:
        totalAn.append((na + na_2 + na_3)/3.0)

# # Artur
fig1 = plt.figure()
plt.plot( 'dist', 'time1ArturMedia', label = '1', data=df, marker='o', markerfacecolor='blue', markersize=8, color='skyblue', linestyle = 'dashed', linewidth=2)
plt.plot( 'dist', 'time2ArturMedia', label = '2', data=df, marker='o', markerfacecolor='red', markersize=8, color='red',linestyle = 'dashdot', linewidth=2)
plt.plot( 'dist', 'time3ArturMedia', label = '3', data=df, marker='o', markerfacecolor='green', markersize=8, color='green', linewidth=2)
plt.plot('dist' , 'vmat', data=df, label = 'esperado', color = 'black', zorder = 5, marker='o', markerfacecolor='gray', markersize=8)
fig1.suptitle('Corrida do Artur', fontsize= 15)
plt.xlabel('Distância (m)', fontsize=10)
plt.ylabel('Tempo (s)', fontsize=10)
plt.legend()
plt.grid()
plt.show()
# # # Pedro
fig2 = plt.figure()
plt.plot( 'dist', 'time1PedroMedia', label = '1', data=df, marker='o', markerfacecolor='blue', markersize=8, color='skyblue',linestyle = 'dashed', linewidth=2)
plt.plot('dist' , 'vmpt', data=df, label = 'esperado', color = 'black', zorder = 5, marker='o', markerfacecolor='green', markersize=8)
plt.plot( 'dist', 'time2PedroMedia', label = '2', data=df, marker='o', markerfacecolor='red', markersize=8, color='red',linestyle = 'dashdot', linewidth=2)
plt.plot( 'dist', 'time3PedroMedia',zorder = 1, label = '3', data=df, marker='o', markerfacecolor='green', markersize=8, color='green', linewidth=2)
fig2.suptitle('Corrida do Pedro', fontsize= 15)
plt.xlabel('Distância (m)', fontsize=10)
plt.ylabel('Tempo (s)', fontsize=10)
plt.legend()
plt.grid()
plt.show()
# # # Andrew
fig3 = plt.figure()
plt.plot( 'dist', 'time1AndrewMedia', label = '1', data=df, marker='o', markerfacecolor='blue', markersize=8, color='skyblue',linestyle = 'dashed', linewidth=2)
plt.plot('dist', 'vman', data=df, label = 'esperado', color = 'black', zorder = 5, marker='o', markerfacecolor='green', markersize=8)
plt.plot( 'dist', 'time2AndrewMedia', label = '2', data=df, marker='o', markerfacecolor='red', markersize=8, color='red',linestyle = 'dashdot', linewidth=2)
plt.plot( 'dist', 'time3AndrewMedia', label = '3', data=df, marker='o', markerfacecolor='green', markersize=8, color='green', linewidth=2)
fig3.suptitle('Corrida do Andrew', fontsize= 15)
plt.xlabel('Distância (m)', fontsize=10)
plt.ylabel('Tempo (s)', fontsize=10)
plt.legend()
plt.grid()
plt.show()
```
